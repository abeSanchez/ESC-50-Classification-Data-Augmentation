```python
y, sr = librosa.load('./ESC-50-master/audio/1-101296-A-19.wav', duration=4.67)
ps = librosa.feature.melspectrogram(y=y, sr=sr)
librosa.display.specshow(ps, x_axis='time', y_axis='mel')
plt.colorbar(format='%+2.0f dB')
```

```python
y, sr = librosa.load('./ESC-50-master/audio/1-101296-A-19.wav')
librosa.display.waveplot(y, x_axis='time')
plt.xlabel('Time (seconds)')
plt.ylabel('Amplitude')
plt.title('Original')
plt.tight_layout()
```
